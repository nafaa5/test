## **Text to speech (API ElevenLabs)**

L'objectif de ce code est de convertir du texte en fichier audio en utilisant l'API ElevenLabs. Le code comprend plusieurs étapes clés :

**Configuration et Initialisation** :

Importation des bibliothèques nécessaires (os, uuid, VoiceSettings et ElevenLabs de la bibliothèque elevenlabs, IPython.display).
Initialisation du client ElevenLabs avec une clé API.

**Fonction text_to_speech_file** :

Cette fonction prend un texte en entrée et utilise l'API ElevenLabs pour convertir ce texte en fichier audio.
Les paramètres détaillés de l'API sont configurés, y compris l'ID de la voix, le format de sortie, le modèle utilisé, et les paramètres de la voix.
Le fichier audio généré est sauvegardé avec un nom de fichier unique.
Le fichier audio est ensuite lu automatiquement en utilisant le module IPython.display.

**Utilisation de la Fonction** :

Un exemple d'utilisation de la fonction text_to_speech_file est fourni, où un texte multilingue est converti en fichier audio et lu automatiquement.

Ce code est utile pour les applications nécessitant la conversion de texte en audio, comme les outils d'apprentissage des langues, les systèmes de lecture pour les malvoyants, ou les assistants vocaux.

```python
import os
import uuid
from elevenlabs import VoiceSettings
from elevenlabs.client import ElevenLabs
from IPython.display import Audio, display

ELEVENLABS_API_KEY = os.getenv("ELEVENLABS_API_KEY")
client = ElevenLabs(
    api_key="sk_7698925d3f6e07f63a94e325fff24bb80cb785a1d2653582",
)

def text_to_speech_file(text: str) -> str:
    # Calling the text_to_speech conversion API with detailed parameters
    response = client.text_to_speech.convert(
        voice_id="pNInz6obpgDQGcFmaJgB",  # Adam pre-made voice
        output_format="mp3_22050_32",
        text=text,
        model_id="eleven_turbo_v2_5",  # use the turbo model for low latency
        voice_settings=VoiceSettings(
            stability=0.0,
            similarity_boost=1.0,
            style=0.0,
            use_speaker_boost=True,
        ),
    )

    # Generating a unique file name for the output MP3 file
    save_file_path = f"/content/speech_output.mp3"

    # Writing the audio to a file
    with open(save_file_path, "wb") as f:
        for chunk in response:
            if chunk:
                f.write(chunk)

    print(f"{save_file_path}: A new audio file was saved successfully!")

    # Play the audio
    display(Audio(save_file_path))

    # Return the path of the saved audio file
    return save_file_path

# Example usage
text_to_speech_file("Hello! 你好! Hola! नमस्ते! Bonjour! こんにちは! مرحبا! 안녕하세요! Ciao! Cześć! Привіт! வணக்கம்!")
```
